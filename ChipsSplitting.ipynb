```python
import sys
import re
import csv
startIndex =0
endIndex = 1000000
total_index = 0
with open('data/chips/20191009.csv','rb') as fin:
    while(1):
        fin.seek(startIndex,0)
        #print(fin.tell())
        data = fin.read(endIndex)
        rows=''
        try:
            rows = data.decode('utf-8').split('#==#')[1]
        except UnicodeDecodeError as err:
            print('fail')
            endIndex-=1
            continue
        startIndex += len(rows.encode('utf-8'))+4
        #print(rows.split('\n')[0],startIndex,endIndex)
        if len(rows) <= 10:
            continue
        rows = rows.split('\n')
        date = rows[0]
        for row in rows[1:-1]:
            cols = [date]
            r = re.sub('[ \r]','',row).split(',')
            stockno = r[0]
            cols.extend(r[2:])
            with open('data/chips/chips_{}.csv'.format(stockno),'a') as f:
                writer = csv.writer(f,lineterminator='\n')
                writer.writerow(cols)
        print(date+' -- Written!')
```
